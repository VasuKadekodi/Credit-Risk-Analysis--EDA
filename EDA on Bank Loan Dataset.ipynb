# Data Cleaning
- converting nagetive values to positive values
- converting categorical data to numerical data

# for Application data

```python
# converting nagetive DAYS_BIRTH column to positive value
def age(i):
    return (i/-365)
app_data['Age']=app_data.DAYS_BIRTH.apply(age)
app_data.head()
```

```python
# converting DAYS_REGISTRATION, DAYS_ID_PUBLISH,DAYS_EMPLOYED and DAYS_LAST_PHONE_CHANGE to year
app_data['Registration_in_years'] = app_data.DAYS_REGISTRATION.apply(age) 
app_data['ID_Publish_in_years'] = app_data.DAYS_ID_PUBLISH.apply(age) 
app_data['Experience'] = app_data.DAYS_EMPLOYED.apply(age)
app_data['Last_Phone_Change_in_years'] = app_data.DAYS_LAST_PHONE_CHANGE.apply(age)

app_data.head()
```

```python
# converting categorical columns to numerical columns
app_data['GENDER'] = np.where(app_data.CODE_GENDER == 'M', 1 , 0)
app_data.head()
```

```python
# find out uniques in column NAME_CONTRACT_TYPE
app_data.NAME_CONTRACT_TYPE.unique()
```

```python
# getting categoroies using value_counts
app_data.NAME_CONTRACT_TYPE.value_counts() 
```

```python
# converting categorical columns to numerical columns
app_data['CONTRACT_TYPE'] = np.where(app_data.NAME_CONTRACT_TYPE == 'Cash loans', 1, 0) 
app_data.head()
```

```python
# find out uniques in column FLAG_OWN_CAR and getting categoroies 
print('unique:' , app_data.FLAG_OWN_CAR.unique())
print('value_counts: \n', app_data.FLAG_OWN_CAR.value_counts())
```

```python
# converting categorical columns to numerical columns
app_data.FLAG_OWN_CAR = np.where(app_data.FLAG_OWN_CAR=='N', 1, 0)
app_data.head()
```

```python
# find out uniques in column FLAG_OWN_REALTY and getting categoroies 
print('unique:', app_data.FLAG_OWN_REALTY.unique())
print('value_counts: \n', app_data.FLAG_OWN_REALTY.value_counts())
```

```python
# converting categorical columns to numerical columns
app_data.FLAG_OWN_REALTY = np.where(app_data.FLAG_OWN_REALTY=='Y', 1, 0)
app_data.head()
```

# for previous application data

```python
# converting -ve values to +ve
prev_appl['DAYS_DECISION']=prev_appl['DAYS_DECISION'].abs()
prev_appl['SELLERPLACE_AREA']=prev_appl['SELLERPLACE_AREA'].abs()
prev_appl['DAYS_FIRST_DUE']=prev_appl['DAYS_FIRST_DUE'].abs()
prev_appl['DAYS_LAST_DUE_1ST_VERSION']=prev_appl['DAYS_LAST_DUE_1ST_VERSION'].abs()
prev_appl['DAYS_LAST_DUE']=prev_appl['DAYS_LAST_DUE'].abs()
prev_appl['DAYS_TERMINATION']=prev_appl['DAYS_TERMINATION'].abs()
prev_appl['DAYS_FIRST_DRAWING']=prev_appl['DAYS_FIRST_DRAWING'].abs()
```

# Imputing or dropping missing values

```python
# Checking the percentage of missing values for all the columns
app_data_missing=app_data.isna().mean()*100
app_data_missing
```

```python
# removing columns having more than 30% missing percentage
app_data = app_data.loc[:, app_data_missing <= 30]
app_data.isna().mean()*100
```

```python
app_data.shape
```

```python
prev_appl.isna().mean()*100
```

```python
app_data.drop(app_data[['OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE', 'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE']], axis=1, inplace=True)
app_data.head()
```

```python
app_data.shape
```

```python
# We will remove unwanted columns from this dataset

unwanted=['FLAG_MOBIL','FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE', 'FLAG_PHONE',
       'FLAG_EMAIL', 'CNT_FAM_MEMBERS', 'REGION_RATING_CLIENT','REGION_RATING_CLIENT_W_CITY','FLAG_DOCUMENT_3', 'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5',
       'FLAG_DOCUMENT_6', 'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8',
       'FLAG_DOCUMENT_9', 'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11',
       'FLAG_DOCUMENT_12', 'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14',
       'FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17',
       'FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20',
       'FLAG_DOCUMENT_21',]

app_data.drop(labels=unwanted,axis=1,inplace=True)
```

```python
# DATA imputation on EXT_SOURCE
app_data['EXT_SOURCE'] = app_data.loc[:, 'EXT_SOURCE_2':'EXT_SOURCE_3'].mean(axis=1)
app_data.drop(app_data.loc[:, 'EXT_SOURCE_2':'EXT_SOURCE_3'].columns , axis=1, inplace=True)
app_data.shape
```

```python
# retriving the columns which has any null values
app_data.isna().mean()*100
```

From the above list the columns that have less than 13% missing values are

- AMT_ANNUITY 
- AMT_GOODS_PRICE 
- NAME_TYPE_SUITE 
- EXT_SOURCE

Taking one Field at a time :
- AMT_ANNUITY

```python
# AMT_ANNUITY having very very less missing values, filling this with 0.0
app_data.AMT_ANNUITY.fillna(value=0.0, inplace=True)
```

- AMT_GOODS_PRICE

```python
# filling missing values with its median
app_data.AMT_GOODS_PRICE.fillna(app_data.AMT_GOODS_PRICE.median(), inplace=True)
```

- NAME_TYPE_SUITE

```python
app_data.NAME_TYPE_SUITE.value_counts()
```

```python
# Since we changed categorical value to integer.
# Using mode to impute missing values
app_data.NAME_TYPE_SUITE.fillna(app_data.NAME_TYPE_SUITE.mode()[0], inplace=True)
```

- EXT_SOURCE

```python
# Using median to impute missing values
app_data.EXT_SOURCE.fillna(app_data.EXT_SOURCE.median(), inplace=True)
```
